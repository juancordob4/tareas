7. Creamos una funcion que muestre si el numero es el que deseamos y responda con true o false

8. Creamos una funcion llamada hallar el maximo de muestra el mas grande de los dos numeros

```python
def find_max(a, b):
    if a > b: return a
    else: return b

print(find_max(13, 1))
```

14. Una funcion que recibe unos numeros y calcula el promedio sumando y diviendo la suma total.

```python
def average(numbers):
  return sum(numbers) / len(numbers)

print(average([1, 2, 3, 4, 5]))
```

15. Una funcion para calcular el factorial de numero entero positvo.

```python
def factorial(n):
    resultado = 1
    for i in range(1, n + 1):
        resultado = resultado * i
    return resultado

print(factorial(3))
```

4. Graficar dos lineas en el mismo lugar.

•
Line 1: x = [0, 1, 2, 3], y = [0, 1, 4, 9] (blue, solid line, label “Squared”)
•
Line 2: x = [0, 1, 2, 3], y = [0, 1, 2, 3] (red, dashed line, label “Linear”)

y Añadir una leyenda




```python
import matplotlib.pyplot as plt

fig, ax = plt.subplots()
ax.plot([0, 1, 2, 3], [0, 1, 4, 9], color='blue', linestyle='-',  label='Squared')
ax.plot([0, 1, 2, 3], [0, 1, 2, 3], color='red',  linestyle='--', label='Linear')
ax.legend()
plt.show()
```

5. Crear un chat bar con categorias ["A","B","C] y con los valores [10,20,15]. Añadimos el valor en la parte superior usando la función "plt.text()

```python
import matplotlib.pyplot as plt

categorias = ["A", "B", "C"]
valores = [10, 20, 15]

fig, ax = plt.subplots()
ax.bar(categorias, valores)

for i, v in enumerate(valores):
    ax.text(i, v + 0.3, str(v), ha='center')
plt.show()
```

6. Crear una figura con una fila y 2 columnas como subgrafica, En la izquierda una subgrafica pero con y es igual a x al cuadrado junto a la derecha con y es igual a raiz de x, para los valores de 0,1,2,3 y 4.

```python
import matplotlib.pyplot as plt
import numpy as np

x = [0, 1, 2, 3, 4]
fig, axs = plt.subplots(1, 2)

axs[0].plot(x, [i**2 for i in x])
axs[0].set_title("y = x²")

axs[1].plot(x, np.sqrt(x))
axs[1].set_title("y = √x")

fig.tight_layout()
plt.show()
```

8. Personalizamos una linea de la grafica

```python
import matplotlib.pyplot as plt

x = [1, 2, 3]
y = [2, 4, 6]

fig, ax = plt.subplots()
ax.plot(x, y, linewidth=2, linestyle=':', marker='o', markersize=8)
plt.show()
```

9. Creamos un histograma desde los datos con 5 bins. Seleccionamos el edge color a negro

```python
import matplotlib.pyplot as plt

data = [1, 1, 2, 2, 2, 3, 3, 4, 5, 5, 5, 5]

fig, ax = plt.subplots()
ax.hist(data, bins=5, edgecolor='black')
plt.show()
```

Creamos una grafica de pai para las ventas le añadimos titulo y mostramos los porcentajes.

```python
import matplotlib.pyplot as plt

labels = ["Apples", "Bananas", "Oranges"]
sizes  = [30, 45, 25]

fig, ax = plt.subplots()
ax.pie(sizes, labels=labels, autopct='%1.1f%%')
ax.set_title("VENTAS")
plt.show()
```
